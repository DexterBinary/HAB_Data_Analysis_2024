```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import datetime
```

```python
def pressureToAltitude(pressure):
    try:
        if pressure < 0:
            raise ValueError("Pressure cannot be negative")
        StandardSeaLevelPressure = 873.0110842 #22 August 2024 Fort Summer, NM
        return 44330 * (1 - (pressure/StandardSeaLevelPressure)**(1/5.255))
    except ValueError as e:
        print(e)
        return 0.00
def millisecondsToSeconds(milliseconds):
    return milliseconds/1000    
def smoothData(data, column, threshold):
    data['diff'] = data[column].diff().abs()
    data = data[data['diff'] < threshold]
    data.drop(columns=['diff'], inplace=True)
    return data
def thresholdData(data, column, threshold):
    data.loc[data[column] > threshold, column] = threshold
    return data
```

```python
def combineCSV(inputFolder :str,outputFile :str, addDataFileNumberCollumn : bool = True):
    combinedCSV = pd.DataFrame()

    i=0
    cummulativeTime = 0
    cummulatives = []

    while True:
        filePath = os.path.join(inputFolder, f'data{i}.csv')
        if not os.path.exists(filePath):
            break

        df = pd.read_csv(filePath)

        if addDataFileNumberCollumn:
            df['DataFileNumber'] = i
            
        df['Time'] = df['Time'] + cummulativeTime
        cummulativeTime = df['Time'].iloc[-1]
        cummulatives.append(cummulativeTime)
        combinedCSV = pd.concat([combinedCSV,df],ignore_index=True)
        i+=1

    combinedCSV.columns = ['Time','Temperature','Pressure','X Axis Acceleration','Y Axis Acceleration','Z Axis Acceleration','CO2','TVOC','DataFileNumber']

    combinedCSV.to_csv(outputFile, index=False)
    return cummulatives
```

```python
currentDataFile = '../data/modified/allData.csv'
df = pd.read_csv(currentDataFile)
df['Time'] = df['Time'].apply(millisecondsToSeconds)
df['Altitude'] = df['Pressure'].apply(pressureToAltitude)
df.to_csv('../data/modified/formattedData.csv',index=False)
```

```python
df = pd.read_csv('../data/modified/GPSData.csv')
initialTime = datetime.strptime(df['TIME'].iloc[0],'%H:%M:%S')
for index,row in df.iterrows():
    currentTime = datetime.strptime(row['TIME'],'%H:%M:%S')
    delta = currentTime - initialTime
    df['TIME'].iloc[index] = delta.total_seconds()
df.to_csv('../data/modified/formattedGPSData.csv',index=False)
```
