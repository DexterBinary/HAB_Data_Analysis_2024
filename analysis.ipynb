```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import csv
```

```python
def plotData(data, x, y, title, xlabel, ylabel, save_path):
    plt.scatter(data[x], data[y])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(save_path)
    plt.show()
```

```python
def createCSV(filePath, headers):
    file = open(filePath, mode='w', newline='')
    writer = csv.writer(file)
    writer.writerow(headers)
def appendCSV(filePath, data):
    file = open(filePath, mode='a', newline='')
    writer = csv.writer(file)
    writer.writerow(data)
def sortCSV(filePath,columnName):
    data = pd.read_csv(filePath)
    data.sort_values(by=[columnName], ascending=True,inplace=True)
    data.to_csv(filePath, index=False)
```

```python
def pressureToAltitude(pressure):
    StandardSeaLevelPressure = 874.0269148666753 #20 August 2024 Fort Summer, NM
    return 44330 * (1 - (pressure/StandardSeaLevelPressure)**(1/5.255))
def millisecondsToSeconds(milliseconds):
    return milliseconds/1000    
```

```python
df = pd.read_csv("output.csv")
plotData(df, "Time", "Pressure", "Altitude vs Time", "Time (s)", "Altitude (m)", "AltitudeVsTime.png")
```
